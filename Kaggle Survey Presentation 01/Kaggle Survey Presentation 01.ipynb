```python
df1 = pd.read_csv('../input/multipleChoiceResponses.csv')
df1.dataframeName = 'multipleChoice.csv'
nRow, nCol = df1.shape
```

```python
df1.shape
```

```python
# Incomedf01 = Incomedf01[Incomedf01['Q9_Mean'] != 'null']
df1 = df1[df1['Q1'] !='What is your gender? - Selected Choice']
```

```python
df1.shape
```

```python
# df1['Q9_Mean'].replace('70-80,000',75000,inplace=True)
Incomedf01 = df1.copy(deep = True)
```

```python
Incomedf01['Q9_Mean'] = Incomedf01['Q9']
```

```python
Incomedf01['Q9_Mean'].replace('0-10,000',5000,inplace=True)
Incomedf01['Q9_Mean'].replace('10-20,000',15000,inplace=True)
Incomedf01['Q9_Mean'].replace('20-30,000',25000,inplace=True)
Incomedf01['Q9_Mean'].replace('30-40,000',35000,inplace=True)
Incomedf01['Q9_Mean'].replace('40-50,000',45000,inplace=True)
Incomedf01['Q9_Mean'].replace('50-60,000',55000,inplace=True)
Incomedf01['Q9_Mean'].replace('60-70,000',65000,inplace=True)
Incomedf01['Q9_Mean'].replace('70-80,000',75000,inplace=True)
Incomedf01['Q9_Mean'].replace('80-90,000',85000,inplace=True)
Incomedf01['Q9_Mean'].replace('90-100,000',95000,inplace=True)
Incomedf01['Q9_Mean'].replace('100-125,000',112500,inplace=True)
Incomedf01['Q9_Mean'].replace('125-150,000',137500,inplace=True)
Incomedf01['Q9_Mean'].replace('150-200,000',175000,inplace=True)
Incomedf01['Q9_Mean'].replace('200-250,000',225000,inplace=True)
Incomedf01['Q9_Mean'].replace('250-300,000',275000,inplace=True)
Incomedf01['Q9_Mean'].replace('300-400,000',350000,inplace=True)
Incomedf01['Q9_Mean'].replace('400-500,000',450000,inplace=True)
Incomedf01['Q9_Mean'].replace('500,000+',550000,inplace=True)
```

```python
Incomedf01 = Incomedf01[pd.notnull(Incomedf01['Q9'])]
```

```python
Incomedf01.shape
```

```python
Incomedf01['Q9_Mean'].value_counts()
```

```python
Incomedf01['Q9'].replace('I do not wish to disclose my approximate yearly compensation','null',inplace=True)
```

```python
Incomedf01 = Incomedf01[Incomedf01['Q9'] != 'null']
```

# 2. Age Wise

```python
Incomedf01['Q9_Mean'].value_counts()
```

```python
AgeGroup = Incomedf01.groupby('Q2')
print (AgeGroup['Q9_Mean'].agg(np.size))
```

```python
Incomedf01['Q9_Mean']=Incomedf01['Q9_Mean'].astype(int)
```

```python
print (AgeGroup['Q9_Mean'].agg([np.sum, np.mean,np.size]))
```

## Python matplotlib for showing respondents age and  mean income trend (above sixty people with high income participated in this survey) .
## Age wise increasing trend of average income, statistical correlation shown , but for  above 70 some more analysis to be done as numbers are less those may be pulling average towards outlier figure. In story 2 we will analyze senior citizens group (>59 years)in more details. 

```python
xx=AgeGroup['Q9_Mean'].agg(np.mean)
a1=xx.sort_values(ascending=True)
xticklabels=a1.index
fig = plt.figure(1)
ax = fig.add_subplot(111)
ax.set_xticklabels(xticklabels, rotation = 45)
plt.plot(a1)
```

## 	Age group and mean  income analyzed and tabulated , for age group more than 60 there are less respondents hence those are not taken in this analysis. There is some interesting observations people with older age also have good average income , very high social income groups & elderly people are also responding , aware on ML which is shows penetration level is increasing.



# 3. Education Wise 

```python
EduGroup = Incomedf01.groupby('Q4')
print (EduGroup['Q9_Mean'].agg(np.mean))
```

```python
fig_size = plt.rcParams["figure.figsize"]
fig_size[0] = 15
fig_size[1] = 9

yy=EduGroup['Q9_Mean'].agg(np.mean)
b1=yy.sort_values(ascending=True)
xticklabels=b1.index
fig = plt.figure(1)
ax = fig.add_subplot(111)
ax.set_xticklabels(xticklabels, rotation = 45)
plt.plot(b1)
```

## There is correlation between formal education and income ( here we are excluding people who have not answered or said no school/formal education)
## Here we are showing histogram  and line graph for correlation , as formal education grows mean income also grows.



## Here education level we are converting to number  of years to get correlation coefficient.
